```python
def get_count(sub_s , c):
    count = 0
    for i in range(len(sub_s)):
        if sub_s[i] == c:
            count = count+1
    
    return count

def find(s , c):
    for i in range(len(s)):
        if s[i] == c:
            return i
```

```python
res = []

left = 0
while left<len(s)-1:
    cur = s[left]
    
    count = get_count(s[left+1:] , cur)
    print(count)
    if count == 0:
        res.append(cur)
        left = left+1
        continue
        
    elif count == 1 or count >= 3:
        pos = s[left+1:].find(cur)
        res.append(s[left : left+1+pos+1])
        left = left+1+pos+1
        continue
        
    elif count == 2:
        res.append(s[left:])
        break
        
    left = left+1
    
        
```

```python
for i in range(len(res)):
    print(len(res[i]) , end='')
    if i<len(res)-1:
        print(',' , end='')
```
