```python
def insert_sort(arr):
    length = len(arr)
    count = 0
    
    for i in range(1,length):
        x = arr[i]
        for j in range(i,-1,-1):
            if x < arr[j-1]:
                arr[j] = arr[j-1]
            else:
                count = count + 1
                break
        arr[j] = x
    
    return count
```

```python
r1 = rad.copy()
r2 = rad.copy()

count1 = insert_sort(r1)

r2.reverse()
count2 = insert_sort(r2)

if count1<count2:
    print(count1)
else:
    print(count2)
```
