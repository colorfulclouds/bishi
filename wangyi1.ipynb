```python
from functools import reduce
```

```python
T = int(input())

num = []

for i in range(T):
    num.append(int(input()))
```

```python
def iss(number):
    bi = bin(number)[2:]
    bi_re = reduce(lambda x,y:x+y , list(reversed(bi)))
    
    return bi == bi_re
```

```python


for i in range(T):
    if (iss(num[i])):
        print('YES')
    else:
        print('NO')
```
