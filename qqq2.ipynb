```python
t , k = map(int , input().split(' '))

arr = []

for i in range(t):
    a , b = map(int , input().split(' '))
    arr.append([a , b])
```

```python
arr
```

```python
def pro(num):
    
    if num < k:
        return 1
    
    steps = num//k
    
    count = 1
    for i in range(1,steps+1):
        white = k*i
        red = num-white
        
        count = count+red+1
    
    return count
```

```python
for i in range(t):
    
    count = 0
    for n in range(arr[i][0] , arr[i][1]+1):
        #n
        count = count + pro(n)
    
    print(count)
```
