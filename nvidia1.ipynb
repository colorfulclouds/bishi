```python
import numpy as np
```

```python
x = int(input())
y = int(input())
z = int(input())
```

```python
a = []

for i in range(x):
    tmp = list(map(int , input().split()))
    a.append(tmp)
    
b = []
for i in range(y):
    tmp = list(map(int , input().split()))
    b.append(tmp)
    
```

```python
a
```

```python
b
```

```python
def dot(a_a , b_b):
    res = 0
    
    for i in range(y):
        res += (a_a[i] * b_b[i])
        
    return res
```

```python
def get_col(col):
    res = []
    
    for i in range(y):
        res.append(b[i][col])
        
    return res
```

```python
for i in range(x):
    for j in range(z):
        
        res = dot(a[i] , get_col(j))
        print(res , end=' ')
    
    print()
```
