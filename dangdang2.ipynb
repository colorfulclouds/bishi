```python
num = list(map(int , input().split(',')))
```

```python
def sort():
    left = 0
    right = len(num) - 1
    cur = 0
    while cur<=right:
        if num[cur] == 0:
            num[left] , num[cur] = num[cur] , num[left]
            left = left+1
            cur = cur+1
        elif num[cur] == 1:
            cur = cur+1
        else:
            num[cur] , num[right] = num[right] , num[cur]
            right = right-1
```

```python
sort()
```

```python
for i in range(len(num)):
    print(num[i] , end='')
    if i<len(num)-1:
        print(',' , end='')
```
