```python
n = int(input())

inp = []

for i in range(n):
    tmp = []
    tmp.append(input())
    tmp.append(input())
    
    inp.append(tmp)
```

```python
def issame(mei , bai):
    for i in range(len(mei)):
        if mei[i] != bai[i]:
            return False
        
    return True
```

```python
def foo(new_mei , mei , bai , op):
    res = []
    
    def pro(new_mei , mei , bai , op):
        if len(new_mei) == len(bai):
            if issame(new_mei , bai):
                res.append(op)
                return

        if len(mei) == 0:
            return 


        tmp = mei.pop(0)
        pro(new_mei , mei , bai , op+'d')
        mei.insert(0 , tmp)

        tmp = mei.pop(0)
        new_mei.insert(0 , tmp)
        pro(new_mei , mei , bai , op+'l')
        mei.insert(0 , tmp)
        new_mei.pop(0)

        tmp = mei.pop(0)
        new_mei.append(tmp)
        pro(new_mei , mei , bai , op+'r')
        mei.insert(0 , tmp)
        new_mei.pop()
    
    pro(new_mei , mei , bai , op)
    
    return res
```

```python
res = []
for i in range(n):
    mei = list(map(int , inp[i][0]))
    bai = list(map(int , inp[i][1]))
    
    tmp = foo([] , mei , bai , '')
    if len(tmp) > 1:
        sorted(tmp)
        
    res.append(tmp)
```
