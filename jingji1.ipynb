```python
import numpy as np

y = input() #第一行数据
y_p = input() #第二行数据

y = [int(x) for x in y.split(",")]
y_p = [int(x) for x in y_p.split(",")]

result = 0.0
'''
计算结果放入result中，并打印出来即可print(result)
'''

TP = 0
FN = 0
for i in range(len(y)):
    if y[i] == 1 and y_p[i] == 1:
        TP = TP+1
    elif y[i] == 1 and y_p[i] == 0:
        FN = FN+1

result = TP/(TP+FN)
        
print(result)
```
