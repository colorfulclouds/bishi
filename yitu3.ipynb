```python
from functools import reduce
```

```python
n , m , k = map(int , input().split(' '))
```

```python
rect = []
for i in range(k):
    tmp = []
    for j in range(n):
        tmp = tmp+list(map(int , input().split(' ')))
    
    rect.append(reduce(lambda x,y : str(x)+str(y) , tmp))
```

```python
rect
```

```python
def is_same(tmp , idx):
    one_count = 0
    zero_count = 0
    
    for i in range(k):
        if tmp[i][idx] == '1':
            one_count = one_count+1
        else:
            zero_count = zero_count+1
    
    return one_count , zero_count
    
```

```python
res = 0

for i in range(n*m):
    tmp = sorted(rect, key=lambda x:x[i])
    if is_same(tmp , i):
        res = res+1
```
