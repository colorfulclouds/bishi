```python
s = input()

s = [_ for _ in s]

def isok(s):
    s = ''.join(s)
    if int(s) % 13 == 5:
        return True
    else:
        return False

res = []

def pro(s , idx):
    if idx == len(s):
        if isok(s):
            res.append(1)
            
        return
            
    if s[idx] == '?':
        
        for i in range(10):
            s[idx] = str(i)
            pro(s , idx+1)
        
        s[idx] = '?'
    else:
        pro(s , idx+1)
            
            
pro(s , 0)

print(sum(res))
```
