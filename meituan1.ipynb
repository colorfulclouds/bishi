```python
s = input().split(',')
```

```python
def comp(s1 , s2):
    if len(s1) == 0:
        return 1 #不交换
    
    if len(s2) == 0:
        return -1 #交换
    
    if len(s1)<=len(s2):
        length = len(s1)
    else:
        length = len(s2)
        
    for i in range(length):
        if s1[i] > s2[i]:
            return 1 #交换
        elif s1[i] < s2[i]:
            return -1 #不交换 
    
    if len(s1) <= len(s2):
        return 1 #两者相同 or s1是s2的前缀
    
    if len(s1) > len(s2):
        return -1 #s2是s1的前缀
```

```python
length = len(s)

for i in range(length):
    min_ = i
    for j in range(i,length):
        if comp(s[j] , s[min_]) == 1:
            min_ = j
    
    _ = s[min_]
    s[min_] = s[i]
    s[i] = _
     
for i in range(len(s)):
    print(s[i] , end='')
    
    if i < len(s)-1:
        print(',' , end='')
    
```
