```python
n = int(input())

value = []

for i in range(n):
    value.append(list(map(int , input().split())))
```

```python
if n == 0:
    print(0)
elif n == 1:
    print(max(value[0][0] , value[0][1]))
else:

    dp = [0] * n
    dp[0] = max(value[0][0] , value[0][1])

    dp[1] = max(dp[0]+value[1][0] , value[1][1])

    for i in range(2 , n):
        dp[i] = max(dp[i-1] + value[i][0] , dp[i-2] + value[i][1])
        
    print(dp[-1])
```
