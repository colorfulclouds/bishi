```python
n = int(input())
exp = input().split(' ')
```

```python
def cal(l):
    s=''
    for i in range(len(l)):
        s=s+l[i]
    
    return eval(s)
```

```python
r1 = cal(exp)
```

```python
def process(e):
    for i in range(n-1):
        if int(e[2*i]) > int(e[2*(i+1)]):
            e[2*i] , e[2*(i+1)] = e[2*(i+1)] , e[2*i]
            
            if cal(e) == r1:
                e=process(e)
            else:
                e[2*(i+1)] , e[2*i] = e[2*i] , e[2*(i+1)]
        
    return e
```
