```python
from functools import reduce
```

```python
n = int(input())

T = input()

m = int(input())

s = []

for i in range(m):
    s.append(input())
```

```python
def pro(st):
    s = ""
    
    while len(s) < n:
        s = s+st
    
    clip = reduce(lambda x,y: x+y , list(s)[:n])
    
    return clip == T
```

```python
count = 0

for i in range(m):
    if pro(s[i]):
        count = count+1

print(count)
```
