```python
def get_min_y():
    pos_max = 0
    neg_max = 0
    
    for i in range(n):
        if zz[i][1] > 0:
            if pos_max < zz[i][1]:
                pos_max = zz[i][1]
        elif zz[i][1] < 0:
            if neg_max < abs(zz[i][1]):
                neg_max = abs(zz[i][1])
    
    if pos_max > neg_max:
        return pos_max
    else:
        return neg_max
```
