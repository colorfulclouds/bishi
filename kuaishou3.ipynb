```python
import math
```

```python
n = int(input())
```

```python
dic = [0 , 0 , 1,1,2,1,2,1,3,2]
```

```python
if n>9:
    for i in range(10 , n+1):
        flag = 1
        for j in range(2 , int(math.sqrt(i))+1):
            if i%j == 0:
                flag = 0
                dic.append(dic[j] + dic[i//j])
                break
        if flag == 1:
            dic.append(1)
        
    print(sum(dic))
else:
    print(sum(dic[:n+1]))
```
