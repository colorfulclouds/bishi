```python
nums = list(map(int , input().split(',')))
k = int(input())
```

```python
nums
```

```python
res = 0
start = 0
mul = 1

for i in range(len(nums)):
    mul *= nums[i]
    
    while mul >= k:
        mul = mul//nums[start]
        start += 1
    
    res = res + i-start+1
```
