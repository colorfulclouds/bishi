```python
t = input().split(':')
```

```python
x_a = '0123456789ABCDEFGHIJKLMNOPQRSTUVWXYZ'

x_d = dict([[k , v] for v , k in enumerate(x_a)])
```

```python
def x2d(num , x):
    res = 0
    
    for i in range(len(num)):
        if x_d[num[len(num)-1-i]] >= x:
            raise ValueError
        res += (x**i) * x_d[num[len(num)-1-i]]
        
    return res
```

```python
res = []

for i in range(2 , 100):
    try:
        hour = x2d(t[0] , i)
        if hour > 23:
            break
        
        minite = x2d(t[1] , i)
        if minite > 59:
            break 
            
        res.append(i)
        
    
    except ValueError:
        pass
```

```python
if len(res) == 0:
    print(-1)
else:
    for r in res:
        print(r , end=' ')
```
