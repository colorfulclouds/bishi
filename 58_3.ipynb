```python
m = int(input())
n = int(input())

grid = []

for i in range(m):
    grid.append(list(map(int , input().split(' '))))
```

```python
def pro():
    for i in range(1,m):
        grid[i][0]+= grid[i-1][0]
    
    for j in range(1, n):
        grid[0][j] += grid[0][j-1]
    
    for i in range(1, m):
        for j in range(1, n):
            grid[i][j] += min(grid[i-1][j],grid[i][j-1]) 

    return grid[-1][-1]
```

```python
pro()
```
