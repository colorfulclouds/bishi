```python
import numpy as np
```

```python
N , M = map(int , input().split(' '))
array = []

for i in range(N):
    tmp = map(int , input().split(' '))
    
    array.append(list(tmp))
    
def func():
    counts = 0
    
    for i in range(N):
        for j in range(M):
            if array[i][j]:
                counts = counts+2
                for r,c in ((i-1,j) , (i+1,j) , (i,j-1) , (i,j+1)):
                    if 0<=r<N and 0<=c<M:
                        val = array[r][c]
                    else:
                        val = 0
                    
                    counts = counts+max(array[i][j]-val , 0)
                
    return counts

print(func())
```

```python
H = np.array(array).max()
has_block = np.zeros([N,M,H])

for i in range(N):
    for j in range(M):
        for k in range(H):
            if array[i][j]>H:
                has_block[i,j,k] = 1
```
