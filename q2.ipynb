```python
n , m = map(int , input().split(' '))
a = list(map(int , input().split()))
b = list(map(int , input().split()))

hep = [[0]*n for _ in range(n)]
res = []
v_c = set()

for i in range(n):
    for j in range(n):
        hep[i][j] = (a[i]+b[j])%m
        
for i in range(n):
    max_value=0
    max_index = 0
    
    for j in range(n):
        if j not in v_c and hep[i][j]>max_value:
            max_value = hep[i][j]
            max_index = j
    
    v_c.add(max_index)
    res.append(max_value)

res.sort()

print(''.join(map(str , res[::-1])))
```
