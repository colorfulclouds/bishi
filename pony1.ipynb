```python
n = int(input())

stone = list(map(int , input().split()))
```

```python
res = 0

target = n-1

while target != 0:
    flag = 1
    
    for i in range(0 , target):
        if stone[i] == stone[target]:
            flag = 0
            target = i
            res += 1
            
            break
            
    if flag == 1:
        target -= 1
        res += 1
    
```

```python
print(res)
```
