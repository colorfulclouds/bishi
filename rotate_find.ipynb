```python
def r_f(arr , begin , end , value):
    mid = (begin + end) // 2
    
    if arr[mid] == value:
        return mid
    
    while begin < end:
        mid = (begin + end) // 2
        
        if arr[mid] == value:
            return mid
        
        if arr[begin] == value:
            return begin
        
        if arr[end] == value:
            return end
        
        if arr[begin] < arr[mid]:
            if value > arr[mid]:
                begin = mid + 1
            else:
                end = mid - 1
            
        else:
            if value > arr[mid]:
                begin = mid + 1
            else:
                end = mid - 1
                
    return -1
```

```python
r_f(arr , 0 , len(arr) - 1 , 1)
```

```python
def find(nums,target):
    
    low,high=0,len(nums)-1
    mid=(low+high)//2
    if nums[mid]==target:
        return mid
    while(low<high):
        mid=(low+high)//2
        if nums[mid]==target:
            return mid
        if nums[low]==target:
            return low
        if nums[high]==target:
            return high
        # [2,3,4,5,1]
        #  l   m   h
        if nums[low]<nums[mid]:#如果前半段有序
            if target >nums[mid]:
                low=mid+1
            else:
                high=mid-1
        else:#如果后半段有序[8,9,1,2,3,4,5,6,7] nums[low]>nums[mid],1
              #             l       m       h
            if target >nums[mid]:
                low=mid+1
            else:
                high=mid-1
    return -1
```
