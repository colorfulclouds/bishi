```python
n , a, b, c = map(int , input().split())

m = list(map(int , input().split()))
```

```python
m = [1]+m
```

```python
def argmax(arr):
    max_idx = 0
    
    for i in range(len(arr)):
        if arr[i] > arr[max_idx]:
            max_idx = i
            
    return max_idx
```

```python
res = 0
```

```python
max_idx = argmax(m[1:n]) + 1

if n - m[max_idx] > 0:
    res = res + 1 + (n-m[max_idx]) * c
elif n - m[max_idx] < 0:
    res = res + 1 + (m[max_idx]-n) * b
else:
    res = res + 1
```

```python
while max_idx != 1:
    tmp_max_idx = max_idx
    max_idx = argmax(m[1:max_idx]) + 1
    
    if tmp_max_idx - m[max_idx] > 0:
        res = res + 1 + (tmp_max_idx-m[max_idx]) * c
    elif tmp_max_idx - m[max_idx] < 0:
        res = res + 1 + (m[max_idx]-tmp_max_idx) * b
    else:
        res = res + 1
```

```python
print(res)
```
