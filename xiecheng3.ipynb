```python
import sys

n = int(input())
m = int(input())

jingd = {}

for i in range(m):
    tmp = list(map(int, input().split(' ')))
    
    if tmp[0] not in jingd:
        jingd[tmp[0]] = [[tmp[1], tmp[2]]]
    else:
        jingd[tmp[0]].append([tmp[1], tmp[2]])
        
    if tmp[1] not in jingd:
        jingd[tmp[1]] = [[tmp[0], tmp[2]]]
    else:
        jingd[tmp[1]].append([tmp[0], tmp[2]])

times = sys.maxsize

def recur(node, mark, count, t):
    global jingd
    
    if count == n:
        if node == 0:
            global times
            times = min(times, t)
        return 1
    
    temp = jingd[node]
    
    for i in range(len(temp)):
        if mark[temp[i][0]] == 0:
            mark[temp[i][0]] = 1
            recur(temp[i][0], mark, count+1, t+temp[i][1])
            mark[temp[i][0]] = 0
    return 1

mark = []
for i in range(n):
    mark.append(0)
    
recur(0, mark, 0, 0)
if times == sys.maxsize:
    times = -1
    
print(times)
```
