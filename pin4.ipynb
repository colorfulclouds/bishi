```python
n , m , k = map(int , input().split(' '))
```

```python
row = [i+1 for i in range(m)]
col = [i+1 for i in range(n)]
```

```python
ele = []
ele = ele+row
ele = ele+col
ele.pop(0)

for i in range(m-1):
    for j in range(n-1):
        
        ele.append(row[i+1] * col[j+1])
        
        
```

```python
ele = sorted(ele , reverse=True)
```
