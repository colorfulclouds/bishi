```python
import numpy as np

x = input() # 如 x='0,2,3,5'
y = input() # 如 y='0,5,3,2'

xx = [int(val) for val in x.split(",")]
yy = [int(val) for val in y.split(",")]

result = 0

'''
计算结果放入result中，并打印出来即可print(result)
'''

for i in range(len(xx)):
    result = result+np.abs(xx[i]-yy[i])

print(result)
```
