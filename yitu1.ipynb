```python
import math
a = int(input())

l = list(map(int , input().split(' ')))
```

```python
count = 0
z = []

for i in range(a):
    if l[i] == 0:
        count += 1
    else:
        z.append(count)
        count = 0
    
z.append(count)
```

```python
ans = 0

for i in range(1 , len(z) - 1):
    count = math.ceil(z[i]/2) - 1
    ans+=count

    
if z[0] == 2:
    ans1 = 1
elif z[0] > 2:
    ans1 = math.ceil((z[0]+1)/2) - 1
else:
    ans1 = 1

if z[-1] == 2:
    ans2 = 1
elif z[-1]>2:
    ans2 = math.ceil((z[-1]+1)/2) - 1
else:
    ans2 = 0

print(ans+ans1+ans2)
```

```python
n , m , k = map(int , input().split(' '))

rect = []
for i in range(k):
    tmp = []
    for j in range(n):
        tmp = tmp+list(map(int , input().split(' ')))
    
    rect.append(reduce(lambda x,y : str(x)+str(y) , tmp))
    
def is_same(tmp , idx):
    start = tmp[0][idx]
    
    for i in range(1 , k):
        if start != tmp[i][idx]:
            return True
    
    return False
    
res = 0

for i in range(n*m):
    tmp = sorted(rect, key=lambda x:x[i])
    if is_same(tmp , i):
        res = res+1
```
