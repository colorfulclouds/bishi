```python
[row , col] = list(map(int , input().split()))

mat = []

for i in range(row):
    mat.append(list(map(int , input().split())))
```

```python
res = []

circle = 0

while col > 2*circle and row>2*circle:
    endx = col-1-circle
    endy = row-1-circle
    
    for i in range(circle , endx+1):
        number = mat[circle][i]
        res.append(number)
    
    if circle<endy:
        for i in range(circle+1 , endy+1):
            number = mat[i][endx]
            res.append(number)
        
    if circle<endx and circle<endy:
        for i in range(endx-1 , circle-1 , -1):
            number = mat[endy][i]
            res.append(number)
            
    if circle<endx and circle<endy-1:
        for i in range(endy-1 , circle , -1):
            number = mat[i][circle]
            res.append(number)
            
    circle += 1
```

```python
for i in range(len(res)):
    print(res[i] , end=' ')
```
