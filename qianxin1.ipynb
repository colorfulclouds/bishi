```python
class node():
    def __init__(self, k=None, l=None, r=None):
        self.val = k
        self.left = l
        self.right = r

def listcreattree(root,llist,i):
    if i<len(llist):
        if llist[i] =='-1':
            return None
        else:
            root=node(k=llist[i])
            root.left=listcreattree(root.left,llist,2*i+1)
            root.right=listcreattree(root.right, llist,2*i+2)
            return root
    return root


    
```

```python
def find(root , p , q):
    if root == None:
        return root
    
    if root.val == p or root.val == q:
        return root
    
    left = find(root.left , p , q)
    right = find(root.right , p , q)
    
    if left and right:
        return root
    
    if left == None:
        return right
    
    if right == None:
        return left
```
