```python
n = int(input())

coords = []

for i in range(n):
    tmp = list(map(int , input().split(' ')))
    coords.append(tmp)
```

```python
max_area = 0

x_sort = sorted(coords , key=lambda x:x[0])
y_sort = sorted(coords , key=lambda x:x[1])
```

```python
for i in range(n-1):
    for j in range(i+1 , n):
        area = (x_sort[i][2]-x_sort[i][0]) * (x_sort[i][3]-x_sort[i][1])
        
        if area>max_area:
            max_area = area
```

```python
for i in range(n-1):
    for j in range(i+1 , n):
        area = (y_sort[i][2]-y_sort[i][0]) * (y_sort[i][3]-y_sort[i][1])
        
        if area>max_area:
            max_area = area
```
