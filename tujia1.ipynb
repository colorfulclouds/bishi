```python
n , m = input().split(' ')
```

```python
def desc():
    for i in range(1 , len(n)):
        if n[i] > n[i-1]:
            return False
    
    return True
```

```python
if desc():
    print(n[: -1*int(m) ])
else:
    m = int(m)
    
    mark = [1]*len(n)
    
    i = 1
    
    while m>0:
        cur = int(n[i])
        for j in range(i-1 , -1 , -1):
            if m > 0 and mark[j] == 1 and int(n[j]) < cur:
                mark[j] = 0

                m -= 1

        i += 1
        
    for i in range(len(n)):
        if mark[i] == 1:
            print(n[i] , end='')
```
