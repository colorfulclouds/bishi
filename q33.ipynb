```python
N , T , M = list(map(int , input().split(' ')))
H = list(map(int , input().split(' ')))
```

```python
print(N,T,M , H)
#N只魔物
#T个回合
#M点法力
```

```python
flag = 1
for x in range(1,1000):
    T_count = T
    M_count = M
    
    for i in range(N):
        blood = H[i]
                    
        while blood != 0:

            if T_count < 0 or M_count < 0:
                break

            if blood == 1:
                T_count = T_count-1
                blood = blood - 1
            elif blood%x == 0:
                t = int(blood/x)

                T_count = T_count-t
                M_count = M_count-t
                blood = blood - t*x
            else:
                t = int(blood/x)+1

                T_count = T_count-t
                M_count = M_count-t
                blood = blood - (t-1)*x - blood%x
            
        if blood != 0:
            break
    
    if T_count>=0 and M_count>=0:
        print(x)
        flag = 0
        break
        
if flag == 1:
    print(-1)
```
