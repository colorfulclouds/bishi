```python
def pro():
    res = [[]]*9
    cols = [[]]*9
    rows = [[]]*9

    for i , row in enumerate(ma):
        for j , num in enumerate(row):
            if num != 'X':
                k = (i//3)*3 + j//3

                if num in rows[i] + cols[j] + res[k]:
                    return False

                rows.append(num)
                cols.append(num)
                res.append(num)

    return True
```

```python
if pro():
    print('true')
else:
    print('false')
```
