```python
n , p , q = map(int , input().split(' '))
```

```python
print(n,p,q)
```

```python
def iss(arr):
    #1 zheng
    #0 fan
    
    z_c = 0
    f_c = 0
    
    for i in range(len(arr)):
        if arr[i] == 1:
            z_c = z_c+1
        else:
            f_c = f_c+1
            
    if z_c<p:
        return False
    
    if f_c<q:
        return False
    
    return True
```

```python
def get_z(arr):
    count = 0 
    for i in range(len(arr)):
        if arr[i] == 1:
            count = count+1
        
    return count
```

```python
count = []

def pro(arr , idx , c):
    if idx == len(arr):
        print(arr)
        
        if iss(arr):
            count.append(get_z(arr))
        return 
        
    
    arr[idx] = 1
    pro(arr , idx+1 , c+1)
    arr[idx] = 0
    
    arr[idx] = 0
    pro(arr , idx+1 , c-1)
    arr[idx] = 1
```

```python
pro([0]*n , 0 , 0)
```

```python
fz = sum(count)
fm = len(count)
```
