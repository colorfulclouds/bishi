```python
'''
数字转阿拉伯
'''
```

```python
num = int(input())
```

```python
res = ''

n_t = [1000 , 900 , 500 , 400 , 100 , 90 , 50 , 40 , 10 , 9 , 5 , 4 , 1]
r_t = ['M' , 'CM' , 'D','CD','C','XC','L','XL','X','IX','V','IV','I']

for i in range(len(n_t)):
    while num >= n_t[i]:
        num -= n_t[i]
        res += r_t[i]
        
```
