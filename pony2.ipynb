```python
n , k = map(int , input().split())
```

```python
score = list(map(int , input().split()))
```

```python
def avg(arr):
    return (sum(arr) - max(arr) - min(arr)) / (len(arr)-2)
```

```python
avg([4,5,6,1])
```

```python
res = 0.0

for i in range(n-k+1):
    tmp = avg(score[i : i+k])
    
    if tmp > res:
        res = tmp
```

```python
print(res)
```
