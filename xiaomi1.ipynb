```python
p = list(map(int , input().split(' ')))
```

```python
def pro():
    if len(p)<=1:
        return 0
    
    l1=[0]*len(p)
    l2=[0]*len(p)
    
    s1 = -1*p[0]
    
    for i in range(1 , len(p)):
        s1 = max(s1 , -1 * p[i])
        l1[i] = max(l1[i-1] , s1 + p[i])
        
    s2 = -1*p[0]
    
    for i in range(1 , len(p)):
        s2 = max(s2 , l1[i] - p[i])
        l2[i] = max(l2[i-1] , s2 + p[i])
        
    return l2[-1]
```

```python
print(pro())
```
