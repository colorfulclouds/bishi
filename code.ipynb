```python
import random
```

```python
dic = []
```

```python
def find(idx):
    for i in range(len(dic)):
        if dic[i] == idx:
            return False
    
    return True
```

```python
def get():
    
    count = 0
    
    while True:
        idx = ''.join(random.choice('0123456789') for i in range(11))
        
        
        count = count+1
        
        if count > 100000000000:
            print('number error')
            break
        
        #print(idx)
        
        if find(idx):
            dic.append(idx)
            return idx
```

```python
def put(number):
    for i in range(len(dic)):
        if dic[i] == number:
            dic.remove(number)
            
            break
```
