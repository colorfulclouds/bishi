```python
s = input()
si = s.split(';')
number = list(map(int , si[0].split(',')))
n = int(si[1])
```

```python
print(number)
```

```python
ou = []
ji = []

for i in number:
    if i%2 == 0:
        ou.append(i)
    else:
        ji.append(i)
```

```python
ou = sorted(ou , reverse=True)
ji = sorted(ji , reverse=True)
```

```python
res = ou.copy()

if n>len(ou):
    for i in range(n-len(ou)):
        res.append(ji[i])
elif n<len(ou):
    for i in range(len(ou) - n):
        res.pop()
```

```python
for i in range(n):
    print(res[i] , end='')
    if i<n-1:
        print(',' , end='')
```
