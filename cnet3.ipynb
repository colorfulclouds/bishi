```python
s = input()

num = list(map(int , (''+s[1:-1]).split(',')))
```

```python
def pro():
    max_s = 0
    
    bak = 0
    
    for i in range(len(num)):
        if bak <= 0:
            bak = num[i]
        else:
            bak += num[i]
            
        if bak > max_s:
            max_s = bak
            
    return max_s
```

```python
pro()
```
