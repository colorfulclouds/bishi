```python
m = int(input())
k = int(input())
```

```python
box = [0]*m

box[0] = 2

for i in range(1 , m):
    box[i] = box[i-1]*i + box[i-1] - i
```

```python
res = 0
tmp = k

for i in range(m-2 , -1 , -1):
   #print(box[i])
    tmp = tmp+box[i]
    print(tmp)
    res += tmp
```
