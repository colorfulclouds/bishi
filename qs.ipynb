```python
arr = list(map(int , input().split(' ')))
```

```python
def p(arr , begin , end):
    start = begin-1
    
    final = arr[end]
    
    for i in range(begin , end):
        if arr[i] <= final:
            start += 1
            arr[start] , arr[i] = arr[i] , arr[start]
        
    arr[end] , arr[start+1] = arr[start+1] , arr[end]
    
    return start + 1
```

```python
p(arr , 0 , len(arr)-1)
```

```python
def qs(arr , begin , end):
    if begin < end:
        pos = p(arr , begin , end)
        
        qs(arr , begin , pos-1)
        qs(arr , pos+1 , end)
```

```python
qs(arr , 0 , len(arr)-1)
```
