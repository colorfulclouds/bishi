```python
T = int(input())
```

```python
def pro(num , k):
    if k<len(num):
        for i in range(len(num)):
            max_idx = i
            for j in range(i+1 , i+k+1):
                if num[j] > num[max_idx]:
                    max_idx = j

            if max_idx != i:
                for k in range(max_idx , i , -1):
                    num[k] , num[k-1] = num[k-1] , num[k]
                break
            
        return num
    else:
        num = pro(num , len(num)-1)
        num = pro(num , k-len(num)+1)
        return num
```

```python
res = []

for i in range(T):
    k = int(input())
    
    n = int(input())
    
    num = list(map(int , input().split()))
    
    tmp = pro(num , k)
    
    res.append(tmp)
```

```python
for i in range(T):
    for j in range(len(res[i])):
        print(res[i][j] , end=' ')
        
    print()
```
