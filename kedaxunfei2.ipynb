```python
def isn(s):
    try:
        int(s)
    except ValueError:
        return False
    
    if len(s) == 1:
        return False
    
    return True
```

```python
res = 0

for i in range(len(s)):
    if isn(s[i]):
        if int(s[i][0]) + int(s[i][-1]) > 8:
            res += 1
```

```python
print(res)
```
