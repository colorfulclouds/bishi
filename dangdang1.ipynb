```python
s = input()
```

```python
def isip(ip):
    arr = ip.split('.')
    
    for tmp in arr:
        if len(tmp) == 0:
            return False
        else:
            if int(tmp) > 255:
                return False
        
    return True
```

```python
res = []

for i in range(1,4):
    for j in range(i+1 , i+4):
        for k in range(j+1 , j+4):
            tmp = s[:i]+'.'+s[i:j]+'.'+s[j:k]+'.'+s[k:]
            if isip(tmp):
                res.append(tmp)
```

```python
for i in range(len(res)):
    print(res[i] , end='')
    if i<len(res) - 1:
        print(',' , end='')
```
