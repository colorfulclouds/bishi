```python
s , m = list(map(int , input().split(' ')))

obj = []

for i in range(s):
    tmp = list(map(int , input().split(' ')))
    obj.append(tmp)
```

```python
obj
```

```python
obj = sorted(obj , key=lambda x:x[0])
```

```python
obj = sorted(obj , key=lambda x:x[1] , reverse=True)
```

```python
obj
```

```python
res = 0

for i in range(s):
    if obj[i][0] < m:
        res += obj[i][1]
        m -= obj[i][0]
```
