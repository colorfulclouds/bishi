```python
n = int(input())
node = list(map(int , input().split(' ')))
```

```python
#9 15 27 35 6
```

```python
def divide():
    ou = []
    ji = []
    
    for p in node:
        if p%2 == 0:
            ou.append(p)
        else:
            ji.append(p)
        
    return ou , ji
```

```python
ou , ji = divide()

ou = sorted(ou)
ji = sorted(ji)
```

```python
def before_count(p):
    if p%2 != 0:
        return int(p/2)
    else:
        return int(p/2)-1
```

```python
print(ou , ji)
```

```python
def start(pos):
    return before_count(pos)
```

```python
def end(pos):
    if pos % 2 == 0:
        fi = 100
    else:
        fi = 99
    
    return before_count(fi) - before_count(pos)
```

```python
M = []
X = []

M.append(ou[0])
X.append(start(ou[0]))

M.append(ou[-1])
X.append(end(ou[-1]))

M.append(ji[0])
X.append(start(ji[0]))

M.append(ji[-1])
X.append(end(ji[-1]))


for i in range(len(ou)-1):
    M.append(ou[i])
    X.append(before_count(ou[i+1]) - before_count(ou[i]) - 1)
    
```

```python
for i in range(len(ji)-1):
    M.append(ji[i])
    X.append(before_count(ji[i+1]) - before_count(ji[i]) - 1)
    
```
