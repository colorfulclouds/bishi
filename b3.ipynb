```python
h , w = map(int , input().split(' '))
```

```python
img = []

for i in range(h):
    pix = list(map(int , input().split(' ')))
    img.append(pix)
```

```python
m = int(input())

kernel = []

for i in range(m):
    filt = list(map(float , input().split(' ')))
    kernel.append(filt)
```

```python
def conv(flip , k):
    summ = 0
    
    for i in range(len(flip)):
        for j in range(len(flip[0])):
            summ = summ + flip[i][j]*k[i][j]
            
    return int(summ)
```

```python
def my_slice(arr , x_start , x_end , y_start , y_end):
    result = []
    
    for i in range(x_start , x_end):
        tmp = []
        for j in range(y_start , y_end):
            tmp.append(arr[i][j])
        
        result.append(tmp)
        
    return result
```

```python
result = []

for i in range(h-1):
    tmp = []
    for j in range(w-1):
        tmp.append(conv(my_slice(img , i,i+m , j,j+m) , kernel))
    
    result.append(tmp)
```

```python
for i in range(len(result)):
    for j in range(len(result[0])):
        print(result[i][j] , end=' ')
    print()
```
