```python
from functools import reduce
```

```python
n = int(input())

x = list(map(int , input().split(' ')))
```

```python
res = []
def pro(arr , idx , cho):
    if len(arr) == len(cho):
        res.append(max(cho))
        return
    
    for i in range(arr[idx]):
        cho.append(i+1)
        pro(arr , idx+1 , cho)
        cho.pop()
```

```python
pro(x , 0 , [])
```

```python
prod = reduce(lambda x,y:x*y , x)

print('%.2f' %(sum(res)/prod))
```
