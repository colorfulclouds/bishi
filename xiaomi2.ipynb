```python
arr = list(map(int , input().split(' ')))
```

```python
def pro():
    res = arr[0]
    presum = 0
    
    for i in arr:
        if presum < 0:
            presum = i
        else:
            presum += i
        res = max(presum,res)
        
    return res
```

```python
print(pro())
```
