```python
#include <stdio.h>

int main(void)
{
    int n;
    long res = 0;
    int i;
    int tmp;
    
    scanf("%d" , &n);

    for(i=n ; i>0 ; i=i/2)
    {
        tmp = (i+1)/2;
        res = res + tmp*tmp;
    }
    
    printf("%d\n", res);
    
    return 0;
}
```
