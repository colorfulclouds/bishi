```python
stra = input()
strb = input()
```

```python
def pro():
    n = len(stra)
    m = len(strb)
    
    dp = [[0]*(m+1) for x in range(n+1)]
    
    for i in range(0 , n+1):
        for j in range(0 , m+1):
            if i==0:
                dp[0][j] = j
            elif j == 0:
                dp[i][0] = i
            else:
                dp[i][j] = min(dp[i-1][j]+1 , dp[i][j-1]+1)
                
                if stra[i-1] == strb[j-1]:
                    dp[i][j] = min(dp[i-1][j-1] , dp[i][j])
                else:
                    dp[i][j] = min(dp[i-1][j-1]+1 , dp[i][j])
                         
    return dp[n][m]
```

```python
pro()
```
