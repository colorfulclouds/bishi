```python
for i in range(len(s)):
    if s[i] == 'R':
        pos = i+s[i+1 : ].find('L') + 1
    else:
        pos = len(s) - 1 - ''.join(reversed(s)).find('R') + 1
    
    print(s , pos)
    
    if (pos - i) % 2 == 0:
        res[i] -= 1
        res[pos] += 1
    else:
        if i != pos:
            res[i] -= 1
            res[pos-1] += 1
```
