```python
def unique(b):
    
    for i in range(1 , len(b)):
        for j in range(i):
            if b[i] == b[j]:
                return False
        
    return True
```

```python
res = []
def pro(b , idx , count):
    if idx == len(b):
        
        if unique(b):
            res.append(count)
            return
        else:
            return
    
    count_no = pro(b , idx+1 , count)

    tmp = b[idx]
    b[idx] = tmp+1
    count_plus = pro(b , idx+1 , count+1)
    b[idx] = tmp
    
```

```python
res = []
for step in range(1,len(s)):
    for i in range(len(s)//step):
        res.append(s[i:i+step])
```

```python
max_count = 0

for i in range(len(res)):
    count = 0
    for j in range(i,len(res)):
        if res[j] == res[i]:
            count = count+1
            
    if count>max_count:
        max_count = count
    
print(max_count)
```
