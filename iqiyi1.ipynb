```python
N = int(input())

a = list(map(int , input().split(' ')))
```

```python
def isok(p):
    for i in range(len(a)):
        if a[i] == 0:
            if p[i] > p[i+1] or p[i] == p[i+1]:
                return False
        else:
            if p[i] < p[i+1] or p[i] == p[i+1]:
                return False
            
    return True
```

```python
count = []

def pro(arr, begin, end):
    if begin == end:
        if isok(arr):
            print(arr)
            count.append(1)
    else:
        j = begin
        for i in range(begin, end):
            arr[i], arr[j] = arr[j], arr[i]
            pro(arr, begin + 1, end)
            arr[i], arr[j] = arr[j], arr[i]


```

```python
pro([i+1 for i in range(N)] , 0 , N)
```

```python
print(sum(count)%int(10e9+7))
```
