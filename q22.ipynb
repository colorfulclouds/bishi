```python
n = int( input() )
```

```python
good = list( map( int , input().split(' ') ) )
```

```python
count = 0
```

```python
dp = [0]*(n+2)

for i in range(2 , n+2):
    if good[i-2]+dp[i-2] > dp[i-1]:
        print(good[i-2])
        count = count + 1
        dp[i] = good[i-2]+dp[i-2]
    else:
        dp[i] = dp[i-1]
```

```python
print(dp[n+1] , count-1)
```
