```python
x , y = list(map(int , input().split(' ')))

dp = []

for i in range(y):
    tmp = [0]*x
    
    dp.append(tmp)
    
for i in range(x):
    dp[0][i] = 1
    dp[i][0] = 1

for i in range(1,y):
    for j in  range(1,x):
        dp[i][j] = dp[i-1][j]+dp[i][j-1]
        
print(dp[x-1][y-1])       
```
